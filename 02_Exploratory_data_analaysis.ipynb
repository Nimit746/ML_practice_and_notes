## Data is majorly of 4 types:- 

<ul>
    <li>Nominal</li>
    <li>Ordinal</li>
    <li>Interval</li>
    <li>Ratio</li>
</ul>

<ol>
<li>Nominal Data</li>
<p>
    - This is the datatype which does not have any numerical value in it. <br/>
    - It can neither be measured nor compared, also no arithmatic operations can be performed on it.<br/>
    - The statastical tendency that can be studied for this type of data is <strong>Mode</strong></br>
    - Examples are: Names, Gender, Address, etc.
</p>
<li>Ordinal Data</li>
<p>
    - This is the datatype whose values follows an order.<br/>
    - It can be compared to each other.<br/>
    - The statastical tendency that can be studied for this type of data is <strong>Median</strong></br>
    - Examples are: T-Shirt sizes ( S, M, L, XL), Scale in customer surveys ( Always, Sometimes, Rarely, Never) etc.
</p>
<li>Interval Data</li>
<p>
    - This is the datatype which has the property of the ordinal data and the intevals are equally split. <br/>
    - It supports arithmatic operations such as addition and subtraction.<br/>
    - Examples are:Degrees of ffahraniheit scale of temperature. This means that the temperature difference observed between 30Deg F and 31deg F is the same difference between 99deg F and 100deg F. However we can't say that the 60deg F is twice as hot as 30deg F.
</p>
<li>Ratio Data</li>
<p>
    - This is the datatype which have a natural zero point and supports all the properties of interval data. <br/>
    - It supports arithmatic operations such as multiplication and division, etc.<br/>
    - It has a clear defination of the 0 which denotes none of the attributes being studied.</br>
    - We can study the statistica tendency and the measures of spread like variation for this kind of data.
</p>

</ol>

<h3>Tranformation on the Nominal Data</h3>

<p>
    It helps to tranform the nominal data into the encoded part so that it can be preprocessed further.</br>
    It is usually done using the <strong>One Hot Encoder</strong> present in the <strong>sklearn.preprocessing</strong></br>
    Now we will take the random data and use onehotencoding on it to tranform it into the numbers from categorical format.
    </br>
    The sample data is as follows:-
    </br>

|  Name         | Gender | Age | Degree       |
|---------------|--------|-----|--------------|
| Edward Remirez| Male   | 28  | Bachelors   |
| Arnav Sharma  | Male   | 23  | Masters     |
| Sophia Smith  | Female | 19  | High School |


</p>

```python
import pandas as pd
df = pd.DataFrame(
[
["Edward Remirez","Male",28,"Bachelors"],
["Arnav Sharma","Male",23,"Masters"],
["Sophia Smith","Female",19,"High School"]
], columns=['Name','Gender','Age','Degree']
)
print(df)
```

```python
from sklearn.preprocessing import OneHotEncoder
gender = OneHotEncoder().fit_transform(df[['Gender']])
# gender.categories_
```

```python
# gender_val = gender.transform(df[['Gender']])
gender.toarray()
```

```python
df[['Gender_F','Gender_M']] = gender.toarray()
print(df)
```

<h3>Tranformation on the Ordinal Data</h3>

<p>
    It helps to tranform the Ordinal data into the encoded part so that it can be preprocessed further.</br>
    It is usually done using the <strong>Ordinal Encoder</strong> present in the <strong>sklearn.preprocessing</strong></br>
    Now we will take the random data and use ordinalencoding on it to tranform it into the numbers from categorical format.
    </br>
    The sample data is as follows:-
    </br>

|  Name         | Gender | Age | Degree      | Gender_F | Gender_M |
|---------------|--------|-----|-------------|:--------:|:--------:|
| Edward Remirez| Male   | 28  | Bachelors   | 0.0      | 1.0      |
| Arnav Sharma  | Male   | 23  | Masters     | 0.0      | 1.0      |
| Sophia Smith  | Female | 19  | High School | 1.0      | 0.0      |


</p>

```python
from sklearn.preprocessing import OrdinalEncoder
encoder = OrdinalEncoder()
encoder.fit_transform(df[['Degree']])
encoder.categories_
```

```python
encoder = OrdinalEncoder(categories=[['Masters','High School','Bachelors','Doctoral']])
df[['Degree_encoded']] = encoder.fit_transform(df[['Degree']])
print(df)
```

```python
df.drop(columns=['Gender','Degree','Gender_F','Name'], inplace=True)
print(df)
```
