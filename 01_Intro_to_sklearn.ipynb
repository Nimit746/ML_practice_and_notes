# This is the Notebook for the introduction to the scikit-learn module in python.

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
print(iris)
print("\n")
print(iris.keys)
```

### This dataset contains majorly 7 keys inside of it that are:-

<ul>
    <li> Data </li>
    <li> Frames </li>
    <li> Target </li>
    <li> Target_names </li>
    <li> Feature_names </li>
    <li> Filename </li>
    <li> DESCR </li>
</ul>

### All these contain some data inside them but the most used keys are:-

<ul>
<li> Target </li>
<li> Data </li>
<li> Feature_names </li>
<li> Target_names </li>
</ul>


### Where:- 

<ul>
<li> Data contains <strong>2D numpy array having 150 rows and 4 columns</strong> </li>
<li> Target contains 150 items which have the numbers (0,1 and 2) refering the 3 target_names.</li>
<li> Target_names contains the 3 varieties of the iris flower that are ('Setosa','Versicolor','Virginica')</li>
<li> Feature_names contain the meaning of the four columns that are contained in the <strong>Data</strong> (that is names of the 4 columns)</li>

```python
print(iris.data[:10])
print(iris.target[0:150:51])
print(iris.target_names)
print(iris.feature_names)
```

### Now here we will write a code for the SVM ( Support Vector Machine ) which will be studied later

```python
from sklearn.svm import SVC
clf = SVC(gamma=0.001, C=100.)
clf.fit(iris.data[:-1], iris.target[:-1])
print(f'The prediction is: {clf.predict(iris.data)}')
```
